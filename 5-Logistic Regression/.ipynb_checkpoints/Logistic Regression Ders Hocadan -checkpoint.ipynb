## Train | Test Split and Scaling

```python
X = df.drop("test_result", axis = 1)
y = df["test_result"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

## Modelling 

https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
log_model.fit(X_train_scaled, y_train)
```

```python
df.head(2)
```

```python
log_model.coef_
```

```python
log_model.intercept_
```

```python
y_pred = log_model.predict(X_test_scaled)
y_pred
```

```python
y_pred_proba = log_model.predict_proba(X_test_scaled) # 0 ve 1 klaslarına ait olma olasılık değerlerini verir
y_pred_proba  # yukarıdaki 1 ve 0 sıfır değerleri oluşturan olasılık değeri veriyor 0 klasındaki değer > 0.5 => 0 , değilse 1 klası büyktür 1 yazar
```

## X_test + y_yest + y_pred + y_pred_proba

```python
test_data = pd.concat([X_test, y_test], axis=1)
test_data
```

```python
test_data["pred_proba"] = y_pred_proba[:,1]
test_data
```

```python
test_data["pred"] = y_pred
test_data
```

```python
test_data.sample(10)
```

### PRC and ROC AUC Trick 

ROC Curve ve Precision-Recall Curve skorlarımıza bakarak, datamızın balanced mı unbalanced mı olduğuna dair fikir sahibi olabiliriz. İki skor birbirine yakın ve aynıysa balanced, arada fark varsa unbalanced olabileceğine dair çıkarım yapabiliriz.
