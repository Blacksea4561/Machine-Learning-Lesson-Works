 # Error Metrics

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score # rmse  kütüphanese yok
# bütün modellerimizde  error metrics le  çalışıyoruz.Çok önemli
# modelizin için de outliear var mı?
# modelimizin ezberleme yapıyor mu?
#ben nekadar  doğru  model  kurmuşum
```

```python
r2_score = r2_score(y_test, y_pred)
r2_score
# r2 skoru 1 e yaklaştıkça iyi  olıuken
# mae, mse, rmse  ise  0 a yaklaştıkça iyi oluyor.
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
# residual lerin mutlak değerini alıp toplayımp   gözlem sayısna bölüyoruz,ortalamasını alıyoruz.
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
# mae ile mse arasında  çok fark varsa    hata vardır diyoruz. sorun
#  target test ile  target   pred  farkını yani residul  lerin kaersini  alıp gözlem sayısına  bölüyor.
# ortalamsınız alıyoruz.
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse
# daha objektif bir bakış açısı var  bunu  sıklıkla kullanacağız
# mse nin karekokün alıyoruz
# rmse  küçük hatayı  kendine yaklaştırır, küçültürken (mesela 0,5**2),  büyük hataları   kendinden uzaklaştırır cezalandırır.
#mae  ile rmse  ne kadar yakın ise  o kadar  iyi olur.
# hataları outlierları göstermesi  daha gerçekçi
```

'''RMSE, tahmin edilen değerlerin gerçek değerlerle olan farklarının karelerinin ortalamasının kareköküdür. RMSE, MAE'ye göre daha yüksek değerler alır ve aykırı değerlere daha hassastır. Bu nedenle, RMSE, özellikle regresyon problemlerinde kullanılan bir ölçüttür.

MAE ise, tahmin edilen değerlerin gerçek değerlerle olan farklarının mutlak değerlerinin ortalamasıdır. MAE, RMSE'ye göre daha küçük değerler alır ve aykırı değerlere daha az hassastır. Bu nedenle, MAE, özellikle sınıflandırma problemlerinde kullanılan bir ölçüttür.

RMSE ve MAE arasındaki fark, kullanılan matematiksel formülden kaynaklanmaktadır. RMSE, farkların kareleri ortalamasının kareköküdür, yani büyük hataların cezalandırılması söz konusudur. MAE ise, farkların mutlak değerlerinin ortalamasıdır, yani hataların büyüklüğüne bakılmaksızın tüm hataların eşit bir şekilde cezalandırılmasıdır.'''

# Adjusted R2 Score (düzenlenmiş R2 skor)

```python
# gözlem sayısını artırmada  featrue sayısını zartırırsak yalncı bir iyileşme oluyor
# eğer  datamıza  ekstra   satır  girerse  
# sensitivite açısısndan  onemli
# ilave feature  gelirse  ilave row da gelmeli  firmadan istemeliyiz
```

```python
def adj_r2(y_test, y_pred, df):
    r2 = r2_score(y_test, y_pred)
    n = df.shape[0]   # number of observations gözlem
    p = df.shape[1]-1 # number of independent variables feature -1 ile 
    adj_r2 = 1 - (1-r2)*(n-1)/(n-p-1) # formül  kafa yorma
    return adj_r2
```

```python
adj_r2(y_test, y_pred, df)
```

```python
def adj_r2(y_test, y_pred, df):
    r2 = r2_score(y_test, y_pred)
    n = df.shape[0]   # number of observations
    p = 100 # number of independent variables features
    adj_r2 = 1 - (1-r2)*(n-1)/(n-p-1)
    return adj_r2
```

```python
adj_r2(y_test, y_pred, df)
```

# What is the concept of punishment for RMSE error metric?

```python
variables = {"Actual": [1000, 2000, 3000, 4000, 5000], "pred": [1100, 2200, 3200, 4200, 5300]}
variables2 = {"Actual": [1000, 2000, 3000, 4000, 5000], "pred": [1100, 2200, 3200, 4200, 5800]}
```

```python
variables,variables2
```

```python
df_2 = pd.DataFrame(variables)
df_2
```

```python
df_3 = pd.DataFrame(variables2)
df_3
```

```python
df_2["residual"] = abs(df_2.Actual - df_2.pred)
df_2

df_3["residual"] = abs(df_3.Actual - df_3.pred)
df_3
# abs   mutlak değer  aldık
```

```python
df_2.residual.sum() / 5
```

```python
df_3.residual.sum() / 5
```

```python
((df_2.residual ** 2).sum() / 5 ) **0.5
```

```python
((df_3.residual ** 2).sum() / 5 ) **0.5
```

```python
# rmse    hata küçükse   az ceza veriyor, fakat  hata büyükse   kendinden uzaklaştırıyor
```

```python
'''
Mean Absolute Error (MAE); her bir tahmin ve gerçek değer arasındaki farkın mutlak değerinin ortalamasıdır. Ceza vermiyor.

Root Mean Square Error (RMSE); tahmin ettiği değerler ile gerçek değerler arasındaki farkların karelerinin ortalamasının 
kareköküdür. Büyük hataları daha da büyütürken, küçük hataları daha da küçültür.
MAE ile RMSE birbirine yakınsa model başarılı diyebiliriz.'''
```

# Evaluation Model

```python
def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    print("Model testing performance: ")
    print("---------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
```

```python
eval_metric(y_test, y_pred)
```
