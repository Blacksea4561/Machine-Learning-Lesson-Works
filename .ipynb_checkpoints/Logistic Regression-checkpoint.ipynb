## Train | Test Split and Scaling

```python
X = df.drop("test_result", axis = 1)
y = df["test_result"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
# scale yapıyoruz
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
 scaler = StandardScaler()
    # burayı tekrar izle
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.tranform(X_test)
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

##  Modelling
https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html

```python
from sklearn.linear_model import LogisticRegression # model import  ettik 
# olasılık  hesbı var içinde
# proba  var işin içinde
```

```python
log_model =  LogisticRegression() 
```

```python
log_model.fit(X_train_scaled, y_train)# eğitim yaptırdık
```

```python
df.head(2)
```

```python
log_model.coef_ # ilk yas, ikinci  fiziksel skorun yaşa göre  etkisi  daha çok yaşta ise attkıkça 
# testen kalıyorlar
```

```python
log_model.intercept_
# yaş 0 ken   
```

```python
y_pred = log_model.predict(X_test_scaled)
y_pred
# arkada  log. reg  modeli dönüyor   0  a ve 1  yakınlaştırıyordu
```

```python
y_pred_proba = log_model.predict_proba(X_test_scaled) # 0 ve 1 klaslarına ait olma olasılık değerlerini verir
y_pred_proba  # yukarıdaki 1 ve 0 sıfır değerleri oluşturan olasılık değeri veriyor 0 klasındaki değer > 0.5 => 0 , değilse 1 klası büyktür 1 yazar
```

## X_test + y_yest + y_pred + y_pred_proba

```python
test_data = pd.concat([X_test, y_test], axis=1)
test_data
```

```python
test_data["pred_proba"] = y_pred_proba[:,1]
test_data
```

```python
test_data["pred"] = y_pred
test_data
```

```python
test_data.sample(10)
```

```python
test_data.groupby('test_result').pred.value_counts()
```
